```python


# DB connection
db_url = "postgresql://vaish28:vG4D2gnkah4BsXvYAxYiws0ji7MayOsF@dpg-d0ka3b3e5dus73bju3sg-a.singapore-postgres.render.com/Traffic"
engine = create_engine(db_url)

# Load data
query = "SELECT * FROM traffic_stops"
df = pd.read_sql(query, engine)

# 1. Remove columns that contain only missing values
df.dropna(axis=1, how='all', inplace=True)

# 2. Handle NaNs in specific columns
# Fill missing age with median
if 'driver_age' in df.columns:
    df['driver_age'] = df['driver_age'].fillna(df['driver_age'].median())

# Fill gender with 'Unknown'
if 'driver_gender' in df.columns:
    df['driver_gender'] = df['driver_gender'].fillna('Unknown')

# Drop rows missing stop date or stop time (essential columns)
df.dropna(subset=['stop_date', 'stop_time'], inplace=True)

# For all other object/text columns, fill NaNs with 'Unknown'
object_cols = df.select_dtypes(include='object').columns
df[object_cols] = df[object_cols].fillna('Unknown')

# For Boolean columns, fill NaNs with False
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].fillna(False)

# 3. Optional: Preview cleaned data
print(df.info())
print(df.head())

# 4. Push to PostgreSQL (replace existing table if exists)
df.to_sql('traffic_stops', engine, if_exists='replace', index=False)
```
